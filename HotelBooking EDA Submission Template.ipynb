### Variables Description

**1.	hotel:** Categorical variable indicating the type of hotel (Resort Hotel or City Hotel).

**2.	is_canceled:** Binary variable indicating if the booking was canceled (1) or not (0).

**3.	lead_time:** Number of days between the booking date and the arrival date.

**4.	arrival_date_year:** The year of arrival.

**5.	arrival_date_month:** The month of arrival.

**6.	arrival_date_week_number:** The week number of the year for the arrival date.

**7. arrival_date_day_of_month:** The day of the month for the arrival date.

**8.	stays_in_weekend_nights:** Number of weekend nights (Saturday or Sunday) the guest stayed or booked to stay at the hotel.

**9.	stays_in_week_nights:** Number of weekday nights (Monday to Friday) the guest stayed or booked to stay at the hotel.

**10.	adults:** Number of adults included in the booking.

**11.	children:** Number of children included in the booking.

**12.	babies:** Number of babies included in the booking.

**13.	meal:** Type of meal booked (e.g., Bed & Breakfast, Half board, Full board, Undefined).

**14.	country:** Country of origin of the guest.

**15.	market_segment:** Market segment designation (e.g., Online TA, Offline TA/TO, Groups).

**16.	distribution_channel:** Booking distribution channel (e.g., Direct, Travel Agents).

**17.	is_repeated_guest:** Binary variable indicating if the guest is a repeated guest (1) or not (0).

**18.	previous_cancellations:** Number of previous bookings that were canceled by the guest.

**19.	previous_bookings_not_canceled:** Number of previous bookings that were not canceled by the guest.

**20.	reserved_room_type:** Code of the room type reserved by the guest.

**21.	assigned_room_type:** Code of the room type assigned to the guest.

**22.	booking_changes:** Number of changes or modifications made to the booking.

**23.	deposit_type:** Type of deposit made by the guest (e.g., No deposit, Non-refundable, Refundable).

**24.	agent:**  ID of the travel agency that made the booking.

**25.	company:** ID of the company or entity that made the booking.

**26.	days_in_waiting_list:** Number of days the booking was in the waiting list before it was confirmed.

**27.	customer_type:** Type of booking customer (e.g., Transient, Contract, Group, Transient-Party).

**28.	adr:** Average Daily Rate, calculated by dividing the total booking revenue by the number of staying nights.

**29.	required_car_parking_spaces:** Number of car parking spaces requested by the guest.

**30.	total_of_special_requests:** Number of special requests made by the guest.

**31.	reservation_status:** Categorical variable indicating the final status of the reservation (e.g., Check-Out, Canceled, No-Show).

**32.	reservation_status_date:** Date when the reservation status was last updated.

#### Chart - 3 Group the data by arriaval date month and calculate the average daily rate

````python
# Chart - 3 visualization code
# Group the data by arriaval date month and calculate the average daily rate
monthly_adr = hotel_booking.groupby('arrival_date_month')['adr'].mean().reset_index()

month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
monthly_adr['arrival_date_month'] = pd.Categorical(monthly_adr['arrival_date_month'], categories=month_order, ordered=True)
monthly_adr.sort_values('arrival_date_month', inplace=True)

# Create a line chart of average daily rate by month
plt.figure(figsize=(10, 6))
sns.lineplot(x='arrival_date_month', y='adr', data=monthly_adr)
plt.title('Average Daily Rate by Month')
plt.xlabel('Month')
plt.ylabel('Average Daily Rate')
plt.xticks(rotation=45)
plt.show()
````

##### 1. Why did you pick the specific chart?

. In the case of exploring the relationship between the average daily rate and months, a line chart is suitable for the following reasons:

Temporal Relationship: A line chart is effective for displaying a temporal relationship between variables. In this case, it helps visualize the trend in average daily rate over different months, allowing you to observe any patterns or fluctuations over time.

Continuity of Data: The line chart assumes a continuous relationship between the x-axis (months) and y-axis (average daily rate). It helps illustrate the flow and progression of the average daily rate across the months in a smooth and visually appealing manner.

Comparison of Trends: With a line chart, you can compare the average daily rate trends across different months. It allows you to identify months with higher or lower average daily rates and understand how they compare to one another.

#### Chart - 15 - Pair Plot

````python
# Pair Plot visualization code

# Select a subset of columns for the pair plot
columns = ['lead_time', 'stays_in_weekend_nights', 'stays_in_week_nights', 'adults', 'children']

# Create a pair plot
sns.pairplot(hotel_booking[columns])
plt.title('Pair Plot')
plt.show()
````

##### 1. Why did you pick the specific chart?

The pair plot was chosen as the specific chart because it allows for the visualization of the relationships between multiple variables in a single plot. It is especially useful when exploring the pairwise interactions and correlations between different variables.

The pair plot provides the following benefits:

Comprehensive View: The pair plot displays scatter plots for each combination of variables, allowing for a comprehensive view of the relationships between multiple variables at once. This makes it easier to identify patterns, trends, or potential correlations.

Variable Distributions: The diagonal of the pair plot represents the distribution of each variable using histograms or kernel density estimations. This provides insights into the distribution and spread of individual variables.

Correlation Analysis: By examining the scatter plots, you can visually assess the strength and direction of the relationships between variables. This can help identify potential correlations or associations between variables of interest.

Efficient Exploration: The pair plot allows for quick comparisons and exploration of multiple variables, saving time and effort compared to creating individual scatter plots for each pair of variables.

By using the pair plot, you can gain insights into the interrelationships between different variables and identify any potential patterns or trends that exist within the dataset.

##### 2. What is/are the insight(s) found from the chart?


The insights found from the pair plot visualization can vary depending on the specific dataset and the variables selected for analysis. However, here are some potential insights that can be derived from the chart:

Correlation Strength: By examining the scatter plots, you can identify the strength of the relationships between different variables. If the points in a scatter plot form a clear pattern or trend, it indicates a stronger correlation between the variables. This insight helps in understanding how variables relate to each other.

Correlation Direction: The scatter plots can also reveal the direction of the relationships between variables. For example, if an increase in one variable corresponds to an increase in another variable, it suggests a positive correlation. Conversely, if an increase in one variable corresponds to a decrease in another variable, it suggests a negative correlation. This insight helps in understanding the nature of the relationships between variables.

Outliers and Clusters: The scatter plots may highlight any outliers or clusters within the data. Outliers are data points that deviate significantly from the general trend, while clusters indicate groups of data points that are close to each other. These insights can provide valuable information about unusual or distinct patterns within the dataset.

Variable Distributions: The histograms or kernel density plots on the diagonal of the pair plot provide insights into the distribution of each variable. They can reveal the skewness, central tendency, and spread of the variables. This insight helps in understanding the characteristics and range of each variable.
